# Linkedin jobs analysis

The purpose of this project is to use how basic new knowledge of aws to establish a data base of the job posted on linkedin and make some analysis with sql on it.    

The first part of this project is to collect and clean the comp.
We are going to do that we the first part of this code

```python
import pandas as pd

from os import listdir
```

```python
src_in = "./data_raw/"
src_out ="./data/"
```

```python
def describeTab(data):
    """Give some basic information about a compFrame
    """
    
    print(data.info(), end="\n\n")
    display(data.head())
```

```python
data_to_skip = ['employee_counts.csv', 'benefits.csv', 'job_postings.csv', 'companies.csv', 'salaries.csv', 'uscities.csv']
for fic in listdir(src_in):
    if fic in data_to_skip :
        continue
    data = pd.read_csv(src_in + fic)
    data.to_csv(src_out + fic, header = False)
```

```python
#employee_counts file
emp_count = pd.read_csv(src_in + "employee_counts.csv")
describeTab(emp_count)
```

```python
emp_count['time_recorded'] = pd.to_datetime(emp_count['time_recorded'], unit='s')
emp_count.to_csv(src_out + "employee_counts.csv", index = False, header = False)
```

```python
#Benefits file
benef = pd.read_csv(src_in + "benefits.csv")
describeTab(benef)
```

```python
benef['inferred'] = benef['inferred'].astype('bool')
benef.to_csv(src_out + "benefits.csv", index = False, header = False)
```

```python
#Salaries
salaries = pd.read_csv(src_in + 'salaries.csv')
describeTab(salaries)
```

```python
#We have some missing value on salaries columns
#We are going to approximate the med_salary by the center of min_salary and max_salary
null_med = salaries['med_salary'].isna()

salaries.loc[null_med, 'med_salary'] = salaries.loc[null_med, ['max_salary', 'min_salary']].median(axis=1)
```

```python
assert (salaries['med_salary'].isna().sum() == 0 )
salaries.to_csv(src_out + 'salaries.csv', index = False, header = False)
```

```python
#companies
comp = pd.read_csv(src_in + 'companies.csv')
describeTab(comp)
```

```python
comp.loc[comp['name'].isna(), 'name'] = 'Phoenix'
```

```python
comp.loc[ comp['company_size'].isna(), 'company_size'] = comp['company_size'].median()
comp['company_size'] = comp['company_size'].astype('int')
```

We are going to clean the **state** column of US

```python
comp[comp['country'] == 'US']['state'].unique().shape[0]
```

```python
def is_float(str_val):
    """Test if a value is an number
    """
    try:
        float(str_val)
        return True
    except:
        return False

def delete_bracket(str_val):
    """delete the bracket and its content at the end of a string
       The string must have one peer of bracket
    """
    return str(str_val).split('(')[0]
```

```python

us_cities = pd.read_csv(src_in + "uscities.csv", usecols=['city', 'state_id', 'state_name'], index_col='city')
states = us_cities.groupby(by='state_id')['state_name'].first()
cities = us_cities.index.value_counts(sort=True)
```

```python
comp['state'] = comp['state'].apply(delete_bracket)  
comp['state'] = comp['state'].str.replace('.', '')  
comp['state'] = comp['state'].str.strip()
comp['state'] = comp['state'].str.title()

for i in range(comp.shape[0]) :

    if comp.loc[i, 'country'] == 'US':
        st = str(comp.loc[i, 'state']).upper() #The current state
        ci = str(comp.loc[i, 'city']).title() #The current city
        
        if st in states.index :
            comp.loc[i, 'state'] = states[st]
            
        elif (ci in us_cities.index) and (cities[ci] == 1) :
            comp.loc[i, 'state'] = us_cities.loc[ci]['state_name']

comp['state'] = comp['state'].str.title()
comp['state'] = comp['state'].replace({
          'Ny - New York' : 'New York', 'Newyork' : 'New York',
          'Calif' : 'California', 'Ca - California' : 'California', 'Ca-California' : 'California',
          'Sc - South Carolina' : 'South Carolina',
          'Nc North Carolina' : 'North Carolina',
          'And Texas' : 'Texas',
          'Nj - New Jersey' : 'New Jersey', 'Newjersey': 'New Jersey',
          'United States' : 'Usa',
          'Select A State' : '',
          'Ilinois' : 'Illinois', 'Illinos' : 'Illinois',
          'Washington Dc' : 'Washington'})
```

```python
comp[comp['country'] == 'US']['state'].unique().shape[0]
```

```python
comp.to_csv(src_out + 'companies.csv', index = False, header = False)
```

```python
#Job_posting
jobs = pd.read_csv(src_in + 'job_postings.csv', keep_default_na=False)
describeTab(jobs)
```

```python
#Conversion des champs date
cols_time = ['original_listed_time', 'expiry', 'closed_time', 'listed_time']
jobs['remote_allowed'] = jobs['remote_allowed'] == 1.0
for col in cols_time:
    jobs[col] = pd.to_datetime( pd.to_numeric(jobs[col]), unit='ms')

jobs['med_salary'] = pd.to_numeric(jobs['med_salary'])
jobs['max_salary'] = pd.to_numeric(jobs['max_salary'])
jobs['min_salary'] = pd.to_numeric(jobs['min_salary'])
```

```python
#We have some missing value on salaries columns
#We are going to approximate the med_salary by the center of min_salary and max_salary
null_med_job = jobs['med_salary'].isna()

jobs.loc[null_med_job, 'med_salary'] = jobs.loc[null_med_job, ['max_salary', 'min_salary']].median(axis=1)
```

```python
jobs.to_csv(src_out + 'job_postings.csv', index = False, header = False)
```
